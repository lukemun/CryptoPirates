```python
def weighted_avg(df):
    tot_vol = df["volume"].sum()
    weight = df["price"] * df["volume"]
    return weight.sum()/tot_vol

def exchangeAvgDiff(asks, bids):
    askAvg = weighted_avg(asks)
    bidAvg = weighted_avg(bids)
    return (askAvg - bidAvg)/bidAvg
```

```python
def getBalances(polo):
    wallet = pd.DataFrame(polo.returnCompleteBalances()).astype(float)
    return wallet
```

```python
getBalances(polo)["USDT"]
```

```python
def buy(polo):
    wallet = getBalances(polo)
    orders = polo.returnOrderBook('USDT_BTC')
    asks = pd.DataFrame(orders["asks"], columns=["price", "volume"]).astype(float)
#     bids = pd.DataFrame(orders["bids"], columns=["price", "volume"]).astype(float)
    btc_price = weighted_avg(asks)
    # calc volume of btc we can buy with current usdt
    volume = wallet["USDT"]["btcValue"]
    return polo.buy("USDT_BTC", btc_price, volume)
    
```

```python
def sell(polo):
    wallet = getBalances(polo)
    orders = polo.returnOrderBook('USDT_BTC')
#     asks = pd.DataFrame(orders["asks"], columns=["price", "volume"]).astype(float)
    bids = pd.DataFrame(orders["bids"], columns=["price", "volume"]).astype(float)
    btc_price = weighted_avg(bids)
    # calc volume of btc we can buy with current usdt
    volume = wallet["BTC"]
    return polo.sell("USDT_BTC", btc_price, volume)
```
