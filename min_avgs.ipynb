```python
import pandas as pd
import json
```

```python
with open('data.txt', 'r') as f:
    data = f.read()
```

```python
btc_data = pd.read_json(data)
```

```python
% matplotlib inline
pd.set_option('display.max_rows', 500)
```

```python
btc_data
```

```python
btc_data['min'] = btc_data['date'].apply(lambda x : x.minute + x.hour*60)
```

```python
btc_data['5min'] = btc_data['min']//5
```

```python
btc_data = btc_data.groupby('5min', as_index=False).mean()
```

```python
btc_data.plot('5min', 'rate', figsize=(15, 10))
```
